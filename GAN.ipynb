```python
# Import necessary libraries
import os
from PIL import Image
import numpy as np
import tensorflow as tf
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torch.nn import Module, Sequential, Conv2d, ConvTranspose2d, LeakyReLU, BatchNorm2d, ReLU, Tanh, Sigmoid, BCELoss 
import matplotlib.pyplot as plt
```

```python
# Path to the dataset
dataset_path = 'formatted'

# Load the dataset into memory
dataset = []
for img in os.listdir(dataset_path):
    image = Image.open(os.path.join(dataset_path,img))
    dataset.append(np.asarray(image))
    
dataset=np.asarray(dataset)
#print(dataset[:5])
```

```python
# Normalize the dataset
dataset = (dataset.astype('float32') / 255.0 - 0.5) * 2.0
```

```python
print(dataset[:5])
# values in range of [-0.5, 0.5] 
```

```python
# Reshape the dataset
dataset = np.reshape(dataset, (dataset.shape[0], 1280, 1024, 3))
```

```python
dataset.shape
# 1000 images with a size of 1280x1024 pixels and 3 color channels (RGB)
```

```python
# Shuffle the dataset
np.random.shuffle(dataset) # prevent overfitting
```
